```python
import numpy as np

U = np.arange(1,101)
print(U)
```

```python
M = np.arange(1,43)
H = np.arange(22,75)

print(M)
print(H)
```

```python
# Intersection of two arrays
result = np.intersect1d(M,H)
print(result, result.size)
```

```python
# unique values from array
result = np.union1d(M,H)
print(result, result.size)
```
