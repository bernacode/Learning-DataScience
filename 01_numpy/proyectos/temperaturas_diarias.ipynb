```python
import numpy as np

dias = np.arange(1,31)
temperaturas = np.random.randint(10,35,30)
print(dias)
print(temperaturas)
```

```python
temperatura_media = np.mean(temperaturas)
print("Temperatura media: ", round(temperatura_media,2), "C")

indices = np.where(temperaturas > 30)

dias_superaron = dias[indices]
numeros_de_dias = len(dias_superaron)
print("Dias que tuvieron una temperatura > 30 C: ", dias_superaron, numeros_de_dias)

temperatura_maxima = temperaturas.max()
print("Temperatura maxima: ",temperatura_maxima)

indice_maximo = np.argmax(temperaturas)
print("Dia con la mayor temperatura: ", dias[indice_maximo])
```
