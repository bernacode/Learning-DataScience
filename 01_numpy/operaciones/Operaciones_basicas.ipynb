```python
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])
```

Estas operaciones se realizan en elemento a elemento.

```python
# Suma
suma1 = a+b
suma2 = np.add(a,b)
print(suma1)
print(suma2)
```

```python

# Resta
resta1 = a-b
resta2 = np.subtract(a,b)

print(resta1)
print(resta2)
```

```python
# Multiplicacion
multiplicacion1 = a*b
multiplicacion2 = np.multiply(a,b)
print(multiplicacion1 )
print(multiplicacion2)
```

```python
# Division
division1 = a/b
division2 = np.divide(a,b)

print(division1)
print(division2)
```

```python
#Potencia
potencia1 = a**b
potencia2 = np.power(a,b)

print(potencia1)
print(potencia2)
```

```python
#Modulo (resto)
modulo1 = a%b
modulo2 = np.mod(a,b)

print(modulo1)
print(modulo2)
```

Tambien puedes usar numeros escalares, por ejemplo, multiplicar todos los elementos del array por un 10. 

```python
import numpy as np

a = np.array([1,2,3])

print(a+10)
print(a-10)
print(a*10)
print(a/10)
print(a**2)
```
