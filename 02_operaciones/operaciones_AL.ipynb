```python
import numpy as np

A = np.random.randint(1,11, size=(2,2))
B = np.random.randint(1,11,size=(2,2))

print (A)
print (B)
```

```python
# Multiplicacion Matricial

C = A @ B
E = np.dot(A,B)

print (C)
print (E)
```

```python
# Transpuesta

AT = A.T
BT = B.T

print(AT)
print(BT)
```

```python
# Inversa

C = np.linalg.inv(A)
E = np.linalg.inv(B)

print(C)
print(E)
```

Solo se puede cuando la matriz es cuadrada, es decir, nxn

```python
# Determinante

detA = np.linalg.det(A)
detB = np.linalg.det(B)

print(detA)
print(detB)
```

```python
# Rango (Grado de independencia)

rangoA = np.linalg.matrix_rank(A)
rangoB = np.linalg.matrix_rank(B)

print(rangoA)
print(rangoB)
```

```python
# Traza

trazaA = np.trace(A)
trazaB = np.trace(B)

print(trazaA)
print(trazaB)
```

```python
# Autovalores y autovectores

PA = np.linalg.eig(A)
PB = np.linalg.eig(B)

print(PA)
print(PB)
```
