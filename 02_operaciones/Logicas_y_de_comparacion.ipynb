```python
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])
```

Comparacion de elementos a  elemento, devuelve valores booleanos (True or False)

```python
# Igualdad

print(a == b)
```

```python
# Diferente
print(a != b)
```

```python
# Mayor que 
print(a > b)
```

```python
# Menor que 
print(a < b)
```

```python
#Mayor o igual que
print(a >= b)
```

```python
# Algun vedadero
print(np.any( a > b)) 
```

```python
# Todos verdaderos
print(np.all(a > b))
```

```python
# Condicional
print( np.where(a > b, 1, 0))
```
