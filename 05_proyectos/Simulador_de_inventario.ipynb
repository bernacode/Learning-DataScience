```python
import numpy as np

productos = ["Teclado", "Mouse", "Monitor", "Latop", "Auriculares"]
surcursales = ["Centro", "Norte", "Sur", "Este"]
stock = np.array([[12,15,8,10],[9,18,7,14],[5,11,6,9],[7,10,13,12],
                  [14,8,10,16]])

```

```python
total_disponible = np.array([sum(n) for n in stock])
print(total_disponible)
print(np.vstack([productos,total_disponible]))
```

```python
nS = np.size(surcursales)
nP = np.size(productos)


total_por_producto = np.array([sum(stock[:,i]) for i in range(0,nS)])
print(total_por_producto)
```

```python

# Se vendio un  teclado  en la sucursal del norte
stock[1,1] = stock[1,1] - 1
print(stock)

# Se vendio un producto de cada tipo en la sucursal del centro
stock[:,0] = stock[:,0] - 1
print(" ")
print(stock)

# Se compraron  tres monitores en todas las surcursales 
stock[2, :] = stock[2, :] - 3
print(" ")
print(stock)

# Hubo un aumento del 10% en stocks (Todas las sucursales)
stock = (stock * 1.10).astype(int)
print(" ")
print(stock)

# Condicion
stock[ stock > 15] -= 1

```
