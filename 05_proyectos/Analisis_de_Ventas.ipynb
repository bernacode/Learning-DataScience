```python
import numpy as np

# Días
dias = np.array(["Lun", "Mar", "Mié", "Jue", "Vie", "Sáb", "Dom", "Lun", "Mar", "Mié"])

# Productos vendidos por día
productos_vendidos = np.array([120, 135, 150, 160, 145, 180, 200, 175, 190, 210])

# Precio promedio por producto ($)
precio_promedio = np.array([15.5, 15.0, 14.8, 15.2, 15.3, 15.1, 15.7, 15.6, 15.9, 16.0])

# Coste promedio por producto ($)
coste_promedio = np.array([10.2, 10.0, 9.8, 10.1, 10.3, 10.2, 10.4, 10.1, 10.3, 10.5])
```

1. Calcular ingresos totales por dia

```python
ingresos = productos_vendidos * precio_promedio
print(ingresos)
```

2. Calcular beneficio neto diario

```python
beneficio = (precio_promedio-coste_promedio)*productos_vendidos
print(beneficio)
```

3. Obtener estadisticas basicas

```python
prom = np.mean(productos_vendidos)
maximo = np.max(ingresos)
minimo = np.min(ingresos)

pmax = np.argmax(ingresos)

print("Promedio de vendtas: ", round(prom))
print("Maximo: ", maximo)
print("Minimo: ", minimo)
print("Dia con mayor beneficio: ", dias[pmax])
```
