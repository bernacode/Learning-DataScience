```python
import numpy as np

arr = np.arange(2,10)
print(arr)
```

```python
binario = np.array([bin(n)[2:] for n in arr])
print(binario)
```

```python
def tranformar_binario(decimal):
    if decimal == 0:
        return "0"
    
    binario = ""

    while decimal > 0:
        resto = decimal % 2
        binario = str(resto) + binario
        decimal = decimal // 2
    
    return binario
    
```

```python
binario = np.array(tranformar_binario(n) for n in arr)
print(binario)
```
