Nivel 2 - Arrays bidimensionales (matrices)

```python
import numpy as np

arr = np.arange(1,10)
print(arr)
```

```python
matriz = arr.reshape((3,3)) # A partir de un array crea una matriz 3x3
print(matriz)
```

```python
obten = matriz[1][2]
print(obten)

obten = matriz[1]
print(obten)

obten = matriz[2]
print(obten)
```

```python
print(matriz.sum())
print(np.mean(matriz,axis=1))
print(matriz.max())
print(matriz.min())
```
